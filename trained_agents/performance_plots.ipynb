```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
dqn_rewards = pd.read_csv('performance_data/dqn_rewards.csv')
ac = pd.read_csv('performance_data/ac_reward.csv')
hrl = pd.read_csv('performance_data/hrl_reward.csv')
```

```python
models = plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
ac['Value'].rolling(window=10).mean().plot(alpha=0.8)
dqn_rewards['Value'].rolling(window=10).mean().plot()
plt.legend(['A2C', 'DQN'])
plt.xlabel('Episodes')
plt.ylabel('Rewards')
plt.subplot(1, 2, 2)
hrl['Value'].rolling(window=10).mean().plot(alpha=0.8);
dqn_rewards['Value'].rolling(window=10).mean().plot();
plt.legend(['FuN-Lite', 'DQN'])
plt.xlabel('Episodes')
plt.ylabel('Rewards');
plt.tight_layout()
plt.show()
```

```python
models.savefig('performance_data/model_perf')
```

```python
dqn_ep = pd.read_csv('performance_data/dqn_episode_length.csv')
ac_ep = pd.read_csv('performance_data/ac_episode_length.csv')
hrl_ep = pd.read_csv('performance_data/hrl_episode_length.csv')
```
